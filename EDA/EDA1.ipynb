```python
import pandas as pd      #import lib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("EDA1/store_sales.csv")    #lode data
df.head()
df.tail()

df.info()        #understand data
df.describe()
df.shape

df.isnull().sum()     #data cleaning

df.duplicated().sum()    #check dublicate and remove it
df.drop_duplicates(inplace=True)

df['Amount'].mean()
df['Amount'].median()
df['Category'].mode()
df['Amount'].std()

sns.countplot(x='Gender', data=df)  #gender distribution
plt.title("Gender Distribution")
plt.show()

plt.figure(figsize=(8,5))             #age
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()

category_count = df['Category'].value_counts()    #num of purchase by category
category_count

plt.figure(figsize=(10,5))
sns.countplot(
    x='Category',
    data=df,
    order=df['Category'].value_counts().index
)
plt.xticks(rotation=45)
plt.show()

category_sales = df.groupby('Category')['Amount'].sum()     #Revenue by Category
category_sales.sort_values(ascending=False)
category_sales.sort_values().plot(
    kind='barh',
    figsize=(10,6)
)
plt.title("Revenue by Category")
plt.show()

top_items = df['ItemPurchased'].value_counts().head(10)   #top purchase
top_items
top_items.plot(kind='bar')
plt.title("Top 10 Purchased Items")
plt.show()

season_sales = df.groupby('Season')['Amount'].sum()     #season sales
season_sales
season_sales.plot(kind='bar')
plt.title("Sales by Season")
plt.show()

payment_sales = df.groupby('PaymentMethod')['Amount'].sum()   #payment analysis
payment_sales
payment_sales.plot(kind='bar')
plt.title("Revenue by Payment Method")
plt.show()

sns.scatterplot(
    x='DiscountApplied(%)',
    y='Amount',
    data=df
)
plt.show()          #discount analysis
sns.histplot(
    df['PreviousPurchases'],
    bins=15
)
plt.show()

sns.scatterplot(
    x='PreviousPurchases',
    y='Amount',
    data=df
)
plt.show()     #previous purchase vs amount

rating = df.groupby('Category')['ItemRating'].mean()    #rating analysis
rating.sort_values(ascending=False)
rating.plot(kind='bar')
plt.title("Average Rating by Category")
plt.show()

num_cols = ['Age','Amount','ItemRating','DiscountApplied(%)','PreviousPurchases']
corr = df[num_cols].corr()     #Correlation Analysis
plt.figure(figsize=(8,6))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)
plt.show()
```

```python
import pandas as pd      #import lib
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("menu.csv")     #loading
df.head()
df.tail()

df.info()             #cleaning
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates(inplace=True)

print("Calories Mean:", df['Calories'].mean())                   #Statistics
print("Calories Median:", df['Calories'].median())
print("Calories Mode:",df['Calories'].mode())
print("Calories Standard Deviation:", df['Calories'].std())

print("Total_Fat Mean:", df['Total Fat'].mean())
print("Total_Fat Median:", df['Total Fat'].median())
print("Total_Fat Mode:",df['Total Fat'].mode())
print("Total_Fat Standard Deviation:", df['Total Fat'].std())

print("Sugar Mean:", df['Sugars'].mean())
print("Sugar Median:", df['Sugars'].median())
print("Sugar Mode:",df['Sugars'].mode())
print("Suagr Standard Deviation:", df['Sugars'].std())

print("Protein Mean:", df['Protein'].mean())
print("Protein Median:", df['Protein'].median())
print("Protein Mode:",df['Protein'].mode())
print("Protein Standard Deviation:", df['Protein'].std())
df.describe()

df['Category'].value_counts()      #prodect analysis

import seaborn as sns
sns.countplot(
    y='Category',
    data=df,
    order=df['Category'].value_counts().index
)
plt.show()

category_calories = df.groupby('Category')['Calories'].mean()       #average calories and highest
print(category_calories)
df[['Item','Calories']]\
.sort_values(by='Calories',ascending=False)\
.head(10)

#bar chart
df.groupby('Category')['Calories'].mean()\
.plot(kind='bar')
plt.title("Average Calories by Category")
plt.show()
#hist 
sns.histplot(df['Calories'], bins=20)
plt.title("Distribution of Calories")
plt.show()

category_fat = df.groupby('Total Fat')['Total Fat'].mean()
print(category_fat)
df[['Item','Total Fat']]\
.sort_values(by='Total Fat',ascending=False)\
.head(10)

df.groupby('Total Fat')['Total Fat'].mean()\
.plot(kind='bar')
plt.title("Average Total Fat by Category")
plt.show()
sns.histplot(df['Total Fat'], bins=20)
plt.title("Distribution of Total Fat")
plt.show()

category_sugar = df.groupby('Sugars')['Sugars'].mean()
print(category_sugar)
df[['Item','Sugars']]\
.sort_values(by='Sugars',ascending=False)\
.head(10)

df.groupby('Sugars')['Sugars'].mean()\
.plot(kind='bar')
plt.title("Average Sugars by Category")
plt.show()
sns.histplot(df['Sugars'], bins=20)
plt.title("Distribution of Sugars")
plt.show()

category_protein = df.groupby('Protein')['Protein'].mean()
print(category_protein)
df[['Item','Protein']]\
.sort_values(by='Protein',ascending=False)\
.head(10)

df.groupby('Protein')['Protein'].mean()\
.plot(kind='bar')
plt.title("Average Protein by Category")
plt.show()
sns.histplot(df['Protein'], bins=20)
plt.title("Distribution of Protein")
plt.show()

#heatmap
numeric_cols = ['Calories','Total Fat','Cholesterol','Sodium','Carbohydrates','Sugars','Protein']
corr = df[numeric_cols].corr()
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)
plt.show()
```
