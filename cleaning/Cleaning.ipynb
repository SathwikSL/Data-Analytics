```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("AB_NYC_2019.csv")      #Load the Dataset and check missing value and correct it
print(df.info())
print(df.isnull().sum())
df["reviews_per_month"] = df["reviews_per_month"].fillna(0)

df["name"] = df["name"].fillna("Unknown")
df["host_name"] = df["host_name"].fillna("Unknown")

df["last_review"] = pd.to_datetime(df["last_review"])
# Option 1: Keep NaT        # Option 2:
df["last_review"] = df["last_review"].fillna(pd.Timestamp("1900-01-01"))

print("Duplicates:", df.duplicated().sum())       #Check Duplicate Records
df = df.drop_duplicates()

df["last_review"] = pd.to_datetime(df["last_review"])      #Standardize Data Types
print(df.dtypes)

Q1 = df["price"].quantile(0.25)       #Detect Outliers
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df["price"] < lower) | (df["price"] > upper)]
print("Outliers:", len(outliers))

df = df[(df["price"] >= lower) & (df["price"] <= upper)]    

print(df[df["price"] <= 0])                     #Validate Data Integrity
print(df[df["minimum_nights"] <= 0])
print(df[df["availability_365"] > 365])

df.to_csv("AB_NYC_2019_Cleaned.csv", index=False)       #Save Cleaned Dataset
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Load dataset
df = pd.read_csv("CAvideos.csv")
# Display first 5 rows
print(df[['title', 'views']].head())
print("Rows and Columns:", df.shape)
print(df.columns)
print(df.info())

print(df[df['views'] < 0])
print(df[df['likes'] < 0])
print(df[df['dislikes'] < 0])
print(df[df['comment_count'] < 0])
print(df.dtypes)

print(df.isnull().sum())
df['description'] = df['description'].fillna('No Description Available')
print(df['description'].isnull().sum())

duplicates = df.duplicated().sum()
print("Duplicate Rows:", duplicates)
df = df.drop_duplicates()
print(df['video_id'].duplicated().sum())

df['publish_time'] = pd.to_datetime(df['publish_time'])
df['title'] = df['title'].str.strip()
df['channel_title'] = df['channel_title'].str.strip()
df['category_id'] = df['category_id'].astype(int)

Q1 = df['views'].quantile(0.25)
Q3 = df['views'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[
    (df['views'] < lower) |
    (df['views'] > upper)
]
print("Outliers:", len(outliers))

plt.boxplot(df['views'])
plt.title("Views Outlier Detection")
plt.show()

df = df[
    (df['views'] >= lower) &
    (df['views'] <= upper)
]

df.to_csv("CAvideos_cleaned.csv", index=False)
```
