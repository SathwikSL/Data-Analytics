```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load datasets
apps = pd.read_csv("apps.csv")
reviews = pd.read_csv("user_reviews.csv")

# DATA CLEANING
# Remove duplicates
apps.drop_duplicates(inplace=True)
# Rating
apps['Rating'] = pd.to_numeric(
    apps['Rating'],
    errors='coerce'
)
apps['Rating'].fillna(
    apps['Rating'].median(),
    inplace=True
)
# Reviews
apps['Reviews'] = pd.to_numeric(
    apps['Reviews'],
    errors='coerce'
)
# Installs
apps['Installs'] = (
    apps['Installs']
    .astype(str)
    .str.replace('+', '', regex=False)
    .str.replace(',', '', regex=False)
)
apps['Installs'] = pd.to_numeric(
    apps['Installs'],
    errors='coerce'
)
# Size
apps['Size'] = apps['Size'].astype(str)
apps['Size'] = apps['Size'].replace(
    'Varies with device',
    np.nan
)
apps['Size'] = apps['Size'].str.replace(
    'M',
    '',
    regex=False
)
apps['Size'] = apps['Size'].str.replace(
    'k',
    '',
    regex=False
)
apps['Size'] = pd.to_numeric(
    apps['Size'],
    errors='coerce'
)
# Remove missing categories
apps.dropna(
    subset=['Category'],
    inplace=True
)
# Clean reviews
reviews.dropna(
    subset=['Sentiment'],
    inplace=True
)

# CATEGORY EXPLORATION
plt.figure(figsize=(12,6))
sns.countplot(
    y='Category',
    data=apps,
    order=apps['Category'].value_counts().index
)
plt.title("Distribution of Apps by Category")
plt.show()

# METRICS ANALYSIS
plt.figure(figsize=(8,5))
sns.histplot(
    apps['Rating'],
    bins=20,
    kde=True
)
plt.title("Distribution of Ratings")
plt.show()

top_apps = apps.sort_values(
    by='Installs',
    ascending=False
).head(10)
print("\nTop 10 Most Installed Apps:\n")
print(top_apps[['App', 'Installs']])

plt.figure(figsize=(8,5))
sns.boxplot(
    x='Type',
    y='Rating',
    data=apps
)
plt.title("Free vs Paid App Ratings")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Size',
    y='Rating',
    data=apps,
    alpha=0.5
)
plt.title("Size vs Rating")
plt.show()

# SENTIMENT ANALYSIS

plt.figure(figsize=(8,5))
sns.countplot(
    x='Sentiment',
    data=reviews
)
plt.title("Review Sentiment Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(
    reviews['Sentiment_Polarity'],
    bins=30,
    kde=True
)
plt.title("Sentiment Polarity")
plt.show()

# INTERACTIVE VISUALIZATIONS

plt.figure(figsize=(12,8))
sns.countplot(
    y='Category',
    data=apps,
    order=apps['Category'].value_counts().index
)
plt.title("Apps per Category")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Installs',
    y='Rating',
    data=apps,
    alpha=0.5
)
plt.xscale('log')
plt.title("Ratings vs Installs")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(
    x='Sentiment',
    data=reviews
)
plt.title("User Sentiment Analysis")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(
    apps[['Rating', 'Reviews', 'Installs']]
    .corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Matrix")
plt.show()
```
